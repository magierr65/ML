**Kardynalność**

```python
# import pakietów
import pandas as pd
import numpy as np
```

```python
# wczytanie danych
df = pd.read_csv("titanic.txt", delimiter=",")
df.replace("?", np.nan, inplace=True)
```

```python
# wyświetlenie pierwszych 20 obserwacji
df.head(20)
```

```python
# sprawdzenie kardynalności każdej cechy
for cecha in df.columns:
    ilosc = df[cecha].nunique()
    print(f"Liczba etykiet zmiennej {cecha:10} -> {ilosc}")
```

```python
# wyświetlenie liczby wszystkich pasażerów
print('Liczba wszystkich pasażerów wynosi: {}'.format(len(df.name)))
```



Wyniki otrzymanie w punkcie 1 oraz 2 świadczą o tym, że zmienne 'pclass', 'survived', 'sex', 'sibsp', 'parch' oraz 'embarked' mają niską kardynalność.

    Zmienna 'survived' ma tylko dwie wartości, więc jest zmienną binarną (przeżył lub nie).
    Zmienna 'pclass' ma tylko trzy wartości (1, 2 oraz 3). Jest to zmienna kategoryczna o niskiej kardynalności, która opisuje klasę biletu, co można połączyć bezpośrednio z zamożnością danego pasażera.
    Zmienne 'sex' oraz 'embarked' mają małą liczbę rodzajów, co powoduje, że dobrze nadawałyby się do zakodowania poprzez One-Hot Encoding.

Z kolei zmienne takie jak 'name', 'ticket', 'fare', 'cabin', 'body', 'home.dest' mają bardzo wysoką kardynalność. Powoduje to, że są one dużo bardziej unikalne dla każdego pasażera. Podczas użycia takich danych w naszym modelu może dojść do overfittingu.


```python
# wyświetlenie unikalnych etykiet dla zmiennej mówiącej o kabinie danego pasażera.
Cabin_label = np.array(df.cabin.unique())
print(Cabin_label)
print(f'\nUnikalnych etykiet dla kabin jest: {len(Cabin_label)}')
```

```python
# redukcja liczby cech dla zmiennej cabin
df["CabinReduced"] = df['cabin'].astype(str).str[0]
# wyświetelnie 20 pierwszych obserwacji dla cabin i nowej zmiennej CabinReduced
print(df[['cabin', 'CabinReduced']].head(20))
```

```python
# sprawdzenie kardynalności każdej cechy po utworzeniu nowej zmiennej
for cecha in df.columns:
    ilosc = df[cecha].nunique()
    print(f"Liczba etykiet zmiennej {cecha:10} -> {ilosc}")
```

```python
# sprawdzenie o ile procent zredukowano liczbę obserwacji
cabin_before = df['cabin'].dropna().astype(str)
cabin_after = df['CabinReduced'].dropna().astype(str)

n_przed = len(np.unique(cabin_before))
n_po = len(np.unique(cabin_after))

procent = (1 - n_po / n_przed) * 100

print(f"Zredukowano kardynalność o {procent:.2f} %.")
```



Oryginalna zmienna cabin ma bardzo dużą kardynalność → 187 różnych wartości, przy czym pełny zapis kabiny zawiera szczegółowe informacje, których model nie jest w stanie efektywnie wykorzystać.

Natomiast pierwsza litera kabiny oznacza pokład i jego położenie, co już może być informacją istotną.

Redukcja tej zmiennej powoduje jedynie mniejszą szczegółowość dotyczącą pokładu. Wydaje mi się, że ma to jednak więcej plusów, ponieważ ta zmienna będzie korzystniejsza dla modeli ML.

```python
import pandas as pd
# podstawowe informacje o przetworzonych wcześniej danych
df['CabinReduced'].replace("n", np.nan, inplace=True) # zamiana n na Nan dla spojnosci danych
df.to_csv("titanic_clean.txt", sep="\t", index=False)
df.head(20)
```

```python
df.describe(include='all')
```

```python
# definiowanie zmiennych do analizy
col_name = ['cabin', 'CabinReduced', 'sex']
X = df[col_name]
y = df['survived']
```

```python
# import biblioteki do podziału
from sklearn.model_selection import train_test_split
```

```python
# podział danych na trening i test (argument stratify None)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=105, stratify=None
)
print(X_train.shape, X_test.shape)
print(y_train.value_counts(normalize=True))
print(y_test.value_counts(normalize=True))
```

```python
# podział danych na trening i test (argument stratify jako y)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=105, stratify=y
)
print(X_train.shape, X_test.shape)
print(y_train.value_counts(normalize=True))
print(y_test.value_counts(normalize=True))
```

**Opis parametrów:**

test_size – określa rozmiar zbioru testowego. Może być liczbą zmiennoprzecinkową (procent) lub liczbą całkowitą (dokładna liczba próbek).

random_state – ustawia ziarno generatora liczb losowych. Dzięki temu podział danych będzie powtarzalny przy kolejnych uruchomieniach kodu.

stratify – określa zmienną, według której zachowywane będą proporcje klas w zbiorach treningowym i testowym. Domyślnie None (losowy podział). Jeśli ustawimy stratify=y, podział będzie proporcjonalny do klas w y, co jest szczególnie ważne przy nierównomiernie rozłożonych klasach.


```python
# Sprawdzenie, czy rozkład wartości w kolumnach jest podobny w obu zbiorach
for col in col_name:
    print(f'\n{col}:')
    print('Train:')
    print(X_train[col].value_counts(normalize=True))
    print('Test:')
    print(X_test[col].value_counts(normalize=True))
```

```python
# ponowny podział danych bez kolumny 'cabin' na trening i test (argument stratify jako y)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=105, stratify=y
)
print(X_train.shape, X_test.shape)
print(y_train.value_counts(normalize=True))
print(y_test.value_counts(normalize=True))
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
for col in col_name:
    encoder.fit(X_train[col].astype(str))
    X_train[col + '_map'] = encoder.transform(X_train[col].astype(str))
    X_test[col + '_map'] = encoder.transform(X_test[col].astype(str))
```

```python
# wyświetlenie nowego X_traina
X_train.head(20)
```

```python
# wyświetlenie nowego X_testa
X_test.head(20)
```

Nowe kolumny _map zachowują oryginalne kolumny, dzięki czemu można je porównać przed i po kodowaniu.

    Po użyciu funkcji transform nasze zmienne kategoryczne zostały zakodowane za pomocą cyfr. W kolumnie 'CabinReduced' mieliśmy osiem możliwości: {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T', 'NaN'}, które w kolumnie 'CabinReduced_map' zostały zakodowane jako: -- {0, 1, 2, 3, 4, 5, 6, 7, 8}.
    Podobnie kolumna 'sex' miała dwie zmienne kategoryczne {'female', 'male'}, które w kolumnie 'sex_map' zostały zakodowane jako {0, 1}.

Kodowanie liczbowe ułatwia modelom uczenie się, bo wiele modeli (np. drzewa decyzyjne, regresja) nie pracuje bezpośrednio na stringach.


```python
# sprawdzenie liczby brakujących wartości
print(X_train.isna().sum())
print(X_test.isna().sum())
```

```python
# zastąpienie brakujących wartości zerami
X_train_fillna = X_train.copy()
X_test_fillna = X_test.copy()
X_train_fillna = X_train.fillna(0)
X_test_fillna = X_test.fillna(0)
```

```python
# sprawdzenie liczby brakujących wartości
print(X_train_fillna.isna().sum())
print(X_test_fillna.isna().sum())
```



Inne metody imputacji brakujących wartości:

    fillna('brak danych') - zastępuje wartości NaN w kolumnie jako string 'brak danych'. Może być to przydatne dla zmiennych kategorycznych, gdy chcemy zachować informację o brakach danyuch jako osobną kategorię.

    SimpleImputer(strategy='most_frequent') - zastępuje wartości brakujące tą, która najczęściej wystepuje w danej kolumnie. Będzie działać zarówno dla zmiennych kategorycznych, jak i numerycznych.

    SimpleImputer(strategy='median') - zastępuje wartości brakujące medianą kolumny. Bardzo dobra metoda dla zmiennych liczbowych, gdy rozkład jest skośnych i chcemy uniknąć wpływu wartości odstających.

    KNNImputer - imputuje wartości odstające na podstawie najbliższych sąsiadów w przestrzeni cech. Porównywanie odbywa się na podstawie odległości w przestrzeni cech (np. odległość Euklidesowa). Brakująca wartość jest zastępowana wtedy średnią lub mediną wartości tej cechy wśród k najbliższych sąsiadów. Stosowany jest w kolumnach z brakami danych liczbowych.

    IterativeImputer - metoda iteracyjna, która przewiduje brakujące wartości jednej kolumny na podstawie pozostałych cech, powtarzając proces kilka razy.



```python
# użycie SimpleImputera z 'most_frequent' dla testu
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='most_frequent')
X_train_SI = X_train.copy()
X_test_SI = X_test.copy()
X_train_SI['CabinReduced'] = imputer.fit_transform(X_train[['CabinReduced']]).ravel()
X_test_SI['CabinReduced'] = imputer.transform(X_test[['CabinReduced']]).ravel()
```

**Reprezentatywność zbiorów treningowego i testowego**

    Zbiory powinny być podobnie rozłożone pod względem cech i wartości docelowej, aby model uczył się na danych reprezentatywnych i wyniki były rzetelne.

    Jeżeli niektóre kategorie w zbiorze testowym nie występują w treningowym, model może mieć problem z poprawną predykcją.

**Wpływ imputacji braków i kodowania**

    Imputacja braków pozwala wykorzystać wszystkie dane, zamiast usuwać obserwacje z brakami, co zwiększa reprezentatywność danych.

    Kodowanie zmiennych kategorycznych pozwala modelom pracować z danymi liczbowymi.

    Liczba unikalnych kategorii powinna być zachowana po kodowaniu, aby nie wprowadzać nowych wartości ani nie gubić informacji.

**Znaczenie stratyfikacji przy podziale danych**

Stratyfikacja (stratify=y) zapewnia, że proporcje klas w zmiennej docelowej są zachowane w zbiorze treningowym i testowym. Dzięki temu model nie będzie uczył się na nierównomiernie reprezentowanych klasach, co poprawia jakość predykcji, szczególnie przy niezbalansowanych zbiorach.
